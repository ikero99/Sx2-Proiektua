```python
from spacy_conll import init_parser

nlp = init_parser("en_core_web_sm",
                  "spacy",
                  disable_sbd=True,                                    # disable sentence segmentation
                  include_headers=True)


# Parse a given string
doc = nlp("A cookie is a baked or cooked food that is typically small, flat and sweet. It usually contains flour, sugar and some type of oil or fat.") #test

# Get the CoNLL representation of the whole document, including headers
conll = doc._.conll_str
print(conll)
```

```python
import os, glob, re

# LOAD gold standard sentences for each file from .conllu file 
sentences = {}
for filename in sorted(glob.glob("data/*.conllu")):
  print(filename)
  sentences[filename] = []
  with open(filename, 'r', encoding="utf-8") as file: 
    for line in file:
      matchObj = re.match( r'^# text = (.*)', line)
      if matchObj:
        sentences[filename].append(matchObj.group(1))
        #print(matchObj.group(1))
```

```python

# for each file, process each line separately, and write output
for filename in sorted(sentences.keys()):
  ofilename = filename.replace("conllu", "conllu.output")
  print(ofilename)
  with open(ofilename, 'w', encoding="utf-8") as ofile: 
    for sentence in sentences[filename]:
      doc = nlp(sentence)
      conll = doc._.conll_str
      ofile.write(conll)
      ofile.write('\n')
```
